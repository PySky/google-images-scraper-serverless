Sometimes Lambda might fail while downloading an image.

This script retrieves all image URLs, deletes them and reinserts them again.

```python
from lib.url_persistence_dynamo import URLPersistenceDynamo

url_db = URLPersistenceDynamo()
```

```python
def delete_urls(urls):
    for url in urls:
        url_db.delete(url['url'], batch=True)
    url_db.flush_batch()
    print('Deleted {}'.format(len(urls)))
    
```

```python
all_urls = []
while 1:
    urls = url_db.get_all()
    if len(urls) == 0:
        break
    all_urls.extend(urls)
    delete_urls(urls)
    print(len(all_urls))
```

```python
len(all_urls)
```

```python
for url in all_urls:
    attributes = {
        'category_name': url['category_name'].rstrip('-. ')}
    url_db.put_batch(url['url'], **attributes)
    print('.',end='')
url_db.flush_batch()
```
